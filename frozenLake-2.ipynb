```python
%reset -f
import numpy as np
import torch
from IPython.display import clear_output
from IPython import display
import random
import matplotlib.pylab as plt
import copy
import gym

import torch.nn as nn
import torch.nn.functional as F

from gym.envs.toy_text.frozen_lake import generate_random_map
```

```python
gSize = 4 # grid size
nH = 2.    # desired number of hole: this is probabilistic
pH = 1 - nH/(gSize*gSize) # probability of  grid being a hole

desc = generate_random_map(size= gSize, p= pH)
```

```python
env = gym.make('FrozenLake-v1', desc=desc,map_name=None, is_slippery=False)
env.reset()
plt.imshow(env.render(mode="rgb_array"))
```

```python
def oneHot(s,h=0):
    x = np.zeros(gSize*gSize)
    x[s] = 1
    fArray = np.array(env.desc == b"F").astype("float64")
    hArray = np.array(env.desc == b"H").astype("float64")
    gArray = np.array(env.desc == b"G").astype("float64")
    state_ = np.array([ x.reshape(gSize,gSize), fArray, gArray    ])
    a,b = s// gSize, s%gSize

    W = -1
    F = -1
    G =  1
    
    left = W if b==0 else  F*hArray[a,b-1] + G*gArray[a,b-1]
    right = W if b==gSize-1 else  F*hArray[a,b+1] + G*gArray[a,b+1]
    up = W if a==0 else  F*hArray[a-1,b] + G*gArray[a-1,b]
    down = W if a==gSize-1 else  F*hArray[a+1,b] + G*gArray[a+1,b]
    
    state_ =  state_ + h*np.random.rand(state_.size).reshape(state_.shape)/10.
    return state_, np.array([left, down, right, up])+ h*np.random.rand(4)/10.
```

```python
env = gym.make('FrozenLake-v1', desc=desc,map_name=None, is_slippery=False)
env.reset()
s1, n1 = oneHot(0)
s1 = torch.tensor(s1).float()
n1 = torch.tensor(n1).float()

class  MyModel(nn.Module):

    def __init__(self):
        super(MyModel, self).__init__()
        self.mainLayer = nn.Sequential(
                nn.Conv2d(3,12,kernel_size=3,stride= 1,padding=1),
                nn.ReLU(),
                nn.Conv2d(12,24,kernel_size=3,stride=1,padding=1),
                nn.ReLU(),
                nn.Conv2d(24,36,kernel_size=3,stride=1,padding=1),
                nn.ReLU(),            
                nn.Flatten(start_dim=-3,end_dim=-1),
                nn.Linear(576,1200),
                nn.ReLU(),
                nn.Linear(1200,64),
                nn.ReLU(),
                nn.Linear(64,4),
        )
        
        self.sideLayer = nn.Sequential(
            #nn.Flatten(start_dim=-3,end_dim=-1),
            nn.Linear(4,64),
            nn.ReLU(),
            nn.Linear(64,250),
            nn.ReLU(),
            nn.Linear(250,32),
            nn.ReLU(),
            nn.Linear(32,4),
        )

    def forward(self, x1,x2):
        x1 = self.mainLayer(x1)
#         x2 = self.sideLayer(x2)
        
#         x = torch.cat([x1,x2],dim=-1)
#         x = nn.Linear(32,4)(x2)
        return x1

model = MyModel()

# print(state_.shape)
model(
        s1,n1
      #   torch.stack([s1,s1]),
      # torch.stack([n1,n1])
     )
```

```python
model2 = copy.deepcopy(model)
model2.load_state_dict(model.state_dict())

loss_fn = torch.nn.MSELoss()
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

gamma = 0.9
epsilon = 0.3
```

```python
from collections import deque
import sys

epochs = 5001
#losses = []
mem_size = 1000
batch_size = 200
replay = deque(maxlen=mem_size)
max_moves = 50
h = 0
sync_freq = 500
j = 0

for i in range(epochs):
    env = gym.make('FrozenLake-v1', 
                   desc=generate_random_map(size= gSize, p= pH),
                   map_name=None, is_slippery=False)
    env.reset()
    state_, n1 = oneHot(0,h=1)
    state1 = torch.tensor(state_).float()
    n1 = torch.tensor(n1).float()

    status = 1
    mov = 0
    ss1 = 0
    while (status == 1):
        j += 1
        mov += 1
        qval = model(state1,n1)
        qval_ = qval.data.numpy()
        action = np.argmax(qval_) if random.random() < epsilon else np.random.randint(0,4)
        ss2, reward, done, _ = env.step(action)
        #modify the reward
        if done:
            reward = 10 if ss2 == gSize*gSize-1 else -10
        elif ss2 == ss1: # going out of grid
            reward = -1
            done = True
        else:
            reward = -1
        ss1 = ss2
        
        
        state_, n2 = oneHot(ss2,h=1)
        state2 = torch.tensor(state_).float()
        n2 = torch.tensor(n2).float()
        exp = ((state1,n1), action, reward, (state2,n2), done)
        replay.append(exp)
        state1 = state2
        n1 = n2
        
        if len(replay) > batch_size:
            minibatch = random.sample(replay, batch_size)
            state1_batch = torch.stack([s1 for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            n1_batch     = torch.stack([n1 for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            action_batch = torch.Tensor([a for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            reward_batch = torch.Tensor([r for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            state2_batch = torch.stack([s2 for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            n2_batch     = torch.stack([n2 for ((s1,n1),a,r,(s2,n2),d) in minibatch])
            done_batch   = torch.Tensor([d for ((s1,n1),a,r,(s2,n2),d) in minibatch])  
            Q1 = model(state1_batch,n1_batch)
            with torch.no_grad():
                Q2 = model2(state2_batch,n2_batch)
            Y = reward_batch + gamma*((1-done_batch)*torch.max(Q2,dim=1)[0])
            X = Q1.gather(dim=1, index=action_batch.long().unsqueeze(dim=1)).squeeze()
            loss = loss_fn(X,Y.detach())
            # print(i,loss.item())
            # clear_output(wait=True)
            optimizer.zero_grad()
            loss.backward()
            losses.append(loss.item())
            optimizer.step()      
        
        
            if j % sync_freq == 0:
                model2.load_state_dict(model.state_dict())
        if done or mov > max_moves:
            status = 0
            mov = 0
    if i %1000 == 0:
        torch.save(model, "frozenLake-2Models/model_frozenLake_"+str(i) +".pth")
#losses = np.array(losses)
print("done")
```
